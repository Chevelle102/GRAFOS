# 1. Cargar la red social del club de kárate (0,5 puntos)

Pista: buscad entre los generadores de NetworkX.


```python
import networkx as nx
import matplotlib.pyplot as plt


G = nx.karate_club_graph()
```

# 2. Dibujar la red social del club de kárate  (0,5 puntos)

```python
nx.draw_circular(G, with_labels=True)
plt.show()
```

# 3. Datos básicos (0,5 puntos)
1. Número de nodos y número de aristas.
2. Densidad.
3. Diámetro.
4. Grado máximo, mínimo y promedio.

```python
print("Numero de Nodos: " + str(len(G.nodes())))
#print("***************************")
print("Numero de Aristas: " + str(len(G.edges())))
#print("***************************")
print("Densidad: %s" % nx.density(G))
#print("***************************")
print("Diámetro: %d" % nx.diameter(G))
print("***************************")
print("El grafo G tiene los grados: ",nx.degree(G))
print("***************************")
```

```python
import numpy as np
z = list(nx.degree(G))
lista=[]
for grado in z:    
    lista.append(grado[1])

print("Máximo:",np.max(lista))
print("Mínimo:",np.min(lista))
print("Promedio:",np.average(lista))
```

# 4. Identifica al instructor y al administrador. (0,5 puntos)

Son los dos nodos de mayor grado.

```python
print('Instructor:',sorted(G.degree, key=lambda x: x[1], reverse=True)[0])
print('Administrador:',sorted(G.degree, key=lambda x: x[1], reverse=True)[1])
```

# 5. Vuelve a dibujar el grafo pero, esta vez, resaltando los nodos del instructor y del administrador. (0,5 puntos)

(Es suficiente con que tengan un color distinto, por ejemplo)

```python
nx.draw_circular(G,node_size=250, with_labels=True)
nx.draw_circular(G,node_size=600,nodelist=[33,0],node_color='b')
plt.show()
```

# 6. Calcula la centralidad de cercanía (closeness) y de intermediación (betweenness). (0,5 puntos)

```python
print("="*50)
print("Closeness")
print("="*50)
print(nx.closeness_centrality(G))
print()
print()
print()
print("="*50)
print("Betweenness")
print("="*50)
print(nx.betweenness_centrality(G))
```

# 7. Estudia el prestigio en la red mediante PageRank. (0,5 puntos)

```python
fila = nx.pagerank(G)

from operator import itemgetter
sorted(fila.items(), key=itemgetter(1),reverse=True)
```

# 8. Obtén el coeficiente de agrupamiento medio y el número de triángulos completos (0,5 puntos)

```python
print("coeficiente de agrupamiento:",nx.average_clustering(G))
print("triángulos                 :",sum(nx.triangles(G).values())/3)
```

# 9. Obtén las comunidades según el algoritmo de Girvan-Newmann (0,5 puntos)

```python
from networkx.algorithms import community

comp = community.girvan_newman(G)
tuple(sorted(c) for c in next(comp))
```
